```python
# Define the URL of the Wikipedia page with the table
url = 'https://footystats.org/england/premier-league'

# Send an HTTP GET request to the URL
page = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if page.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(page.text, 'lxml')
    
    table = soup.find('table', {'class': 'full-league-table table-sort col-sm-12 mobify-table'})

    
    # Initialize empty lists to store the data from each column
    rankings = []
    teams = []
    MPs = []
    Ws = []
    Ds = []
    Ls = []
    GFs = []
    GAs = []
    GDs =[]
    Points =[]
    
    
    # Iterate through the rows of the table
    for row in table.find_all('tr')[1:]:  # Skip the header row (first row)
        columns = row.find_all('td')
        if len(columns) >= 2:
            ranking = columns[0].get_text(strip=True)
            team = columns[2].get_text(strip=True)
            MP = columns[3].get_text(strip=True)
            W = columns[4].get_text(strip=True)
            D = columns[5].get_text(strip=True)
            L = columns[6].get_text(strip=True)
            GF = columns[7].get_text(strip=True)
            GA = columns[8].get_text(strip=True)
            GD = columns[9].get_text(strip=True)
            Point = columns[10].get_text(strip=True)
            
            rankings.append(ranking)
            teams.append(team)
            MPs.append(MP)
            Ws.append(W)
            Ds.append(D)
            Ls.append(L)
            GFs.append(GF)
            GAs.append(GA)
            GDs.append(GD)
            Points.append(Point)
            
    
    # Print or process the scraped data
    for ranking,team, MP, W, D, L, GF, GA, GD, Point in zip(rankings, teams, MPs, Ws, Ds, Ls, GFs, GAs, GDs, Points):
        print(f"Ranking: {ranking},Team: {team}, Match Played: {MP}, Win: {W}, Draw: {D},Lose:{L} Goal For: {GF}, Goals Against: {GA}, Goal Difference: {GD}, Point: {Point}")
else:
    print('Failed to retrieve the webpage.')
```

```python
csv_file_path = "Premier_League_Table"

with open(csv_file_path, mode='w', newline = '') as file:
    csv_writer = csv.writer(file)
    
    header = ["Ranking", "Team", "Match Played", "Win", "Draw", "Lose", "Goal For", "Goals Against", "Goal Difference", "Point"]
    
    csv_writer.writerow(header)
    
    for ranking, team, MP, W, D, L, GF, GA, GD, Point in zip(rankings, teams, MPs, Ws, Ds, Ls, GFs, GAs, GDs, Points):
            csv_writer.writerow([ranking, team, MP, W, D, L, GF, GA, GD, Point])
```
